# Data Creation of Nominal Column named "colors"


```python
import numpy as np
import pandas as pd

np.random.seed(0) # seed will always create the same sample
num_samples=1000

colors=["Red","Green","Blue","Yellow", "grey","brown"]

nominal_data=np.random.choice(colors, num_samples)
# nominal_data
```

```python
dataframe=pd.DataFrame(nominal_data, columns=['colors'])
dataframe
```

# Data Creation of ordinal Column named "satisfaction_level"


```python
satisfaction_level=['low','medium','high']
ordinal_data= np.random.choice(satisfaction_level,size=num_samples)
dataframe['satisfaction_level']=ordinal_data
dataframe.head()
```

# Data creation of Range Data(Quantitative Data) under Income column

```python
np.random.seed(0)
ratio_data=np.random.randint(10000,10000000,num_samples)
dataframe['salary']=ratio_data
dataframe
```
