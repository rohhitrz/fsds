# Data creation for the income

```python
np.random.seed(1)
#right skewed distribution
income_distribution=np.random.gamma(shape=2.0,scale=1000.0,size=1000)

#calculate the skewness

skew_value=skew(income_distribution)
sns.histplot(income_distribution, bins=30,kde=True)
print(income_distribution)
```

# Kurtosis- measure of tailedness of data distribution

```python
data= pd.read_csv("./process_engineering_data_lyst1750493214144.csv")
data.head()
```

# calculate range of temprature column in the above data

```python
print(data)
```

```python
print(data['temperature'])
```

```python
lowest=9083
highest=-8928
for k,v in data['temperature'].items():
    if(v<lowest):
        lowest=v
    if(v>highest):
        highest=v
print(lowest,highest)
print(f'range is {highest-lowest}')
```

# simple/better way to find the range

```python
print(f"range is {data["temperature"].max()-data["temperature"].min()}")
```

# IQR-------> Q3(75%)-Q1(25%)

```python
q1= data['temperature'].quantile(0.25)
q3=data['temperature'].quantile(0.75)
print("temperature_IQR= ", q3-q1)
```
