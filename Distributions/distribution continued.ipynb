```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

#standard normal distribution
portfolio_return=np.random.normal(loc=0,scale=1,size=1000)



#calculate the kurtosis
from scipy.stats import kurtosis
kurt_value=kurtosis(portfolio_return)
print(f"kurtosis: {kurt_value:.2f}")
sns.histplot(portfolio_return,bins=30,kde=True)
plt.title(f'Portfolio Returns Distribution (kurtosis: {kurt_value:.2f})')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.show()
```
