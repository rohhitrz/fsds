# simple random sampling: 100 records randomly

```python
sample_data=data.sample(n=100,random_state=44)
sample_data
```

# stratified  sampling


```python
stratified_sample=data.groupby('Region').apply(lambda x: x.sample(min(len(x),25), random_state=44))
stratified_sample
```

```python
stratified_sample.to_csv('stratified_sample_data.csv')
```

# Cluster Sampling on manufacture data by Batch

```python
manufacturing_data=pd.read_csv('./Manufacturing_Data.csv')
manufacturing_data.head()
```

```python
clusters= manufacturing_data['BatchNumber'].unique()
clusters
```

```python
selected_cluster=np.random.choice(clusters,size=5,replace=False)
```

```python
selected_cluster
```

```python
clustered_sample= manufacturing_data[manufacturing_data['BatchNumber'].isin(selected_cluster)]
print(clustered_sample)
```
